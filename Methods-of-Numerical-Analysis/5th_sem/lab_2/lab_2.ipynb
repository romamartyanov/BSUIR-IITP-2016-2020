### Постановка задачи:
Будем рассматривать дифференциальное уравнение второго порядка.

<h3><center>$y'' + p(x)y' + q(x)y = f(x)$</center></h3>

где $p(x), q(x), f(x)$ − заданные непрерывные на отрезке  $[a, b]$ функции.
Краевой задачей называется задача нахождения решения , удовлетворяющего граничным условиям:

<h3><center>$\begin{cases} y(a) = A \\ y(b) = B \end{cases}$</center></h3>


Если на границах $х = а$ и $х = b$ заданы значения искомой функции $у(а)$, $у(b)$, то такие условия называются граничными условиями первого рода, а задача называется первой краевой задачей для ОДУ.
Граничные условия второго рода:

<h3><center>$\begin{cases} y'(a) = y_a \\ y'(b) = y_b \end{cases}$</center></h3>

Граничные условия третьего рода (если на границах заданы линейные комбинации искомой функции и ее первой производной):

<h3><center>$\begin{cases} y'(a) + \alpha y(a) = y_a \\ y'(b) + \beta y(b) = y_b \end{cases}$</center></h3>


Рассмотрим дифференциальное уравнение  первого порядка $U' + AU = f(x), x \in [a, b]$

Пусть его решением является функция $U=U(x)$.
Разобьем отрезок $[a,b]$ на n равных частей с шагом $h$, то есть построим последовательность ${x_k}$:

<h3><center>$x_0 = a, x_k = x_0 + kh$, $k = \overline{0, n}$, где  $h=\frac{b-a}{n}$</center></h3>

Множество узлов $x_k$ на отрезке $[a,b]$ образует сетку  с шагом $h$.
При этом, любую функцию $g_k, k=0, 1,…,n$ , определенную на сетке, будем называть сеточной функцией. В частности, сеточной  функцией будет последовательность  порожденная решением  дифференциаль-ного уравнения.

Заменим производные функции во внутренних узлах их разностными аппроксимациями:

<h3><center>$U'(x_k) \approx \frac{U(x_k + h) - U(x_k}{h}$</center></h3>

Значение функции $U$ в $k$–ом узле обозначим

<h3><center>
$L_h U^{(h)} = \begin{cases} a_k U_{k-1} + b_k U_{k} +  c_k U_{k+1}, k = \overline{1, n-1}\\ U_0 \\ U_n\end{cases}$
</center></h3>

<h3><center>
$U(x_k) = U_k$
</center></h3>

<h3><center>
$||U^{(h)}||_{U_k} = ||U^{(h)}|| = \underset{k = \overline{0, n}}{max} |U_k|$
</center></h3>

а функции $f$ соответственно через $f_k = f(x_k)$. В новых обозначениях получим разностное уравнение:

<h3><center>$\frac{U_{k+1} - U_k}{h} + AU_k = f_k$</center></h3>

C большей точностью первую производную можно также представить в виде:

<h3><center>$U'(x_k) \approx \frac{U(x_k + h) - U(x_k - h)}{2h}$</center></h3>

тогда уравнение будет иметь вид

<h3><center>$\frac{U_{k+1} - U_{k-1}}{2h} + AU_k = f_k$</center></h3>

Рассмотрим теперь на данном отрезке дифференциальное уравнение
второго порядка

<h3><center>$U'' + AU' + BU = f(x)$</center></h3>

Аналогично, заменив производные их разностными аппроксимациями, получим:

<h3><center>$\frac{U_{k+1} + 2U_{k} + U_{k-1}}{h^2} + A \frac{U_{k+1} + U_{k-1}}{2h} + BU_k = f_k$</center></h3>

Преобразуем полученное разностное уравнение:

<h3><center>$(\frac{A}{2h} + \frac{1}{h^2}) U_{k+1} + (B - \frac{2}{h^2}) U_k + (\frac{1}{h^2} - \frac{A}{2h}) U_{k-1} = f_k$</center></h3>

Полученные разностные уравнения будем также называть разностными схемами для соответствующих дифференциальных уравнений. Таким образом,   мы перешли от дифференциальных  уравнений  к разностным уравнениям вида:

<h3><center>$a_k U_k + b_k U_{k+1} = f_k$ $(1)$</center></h3>

<h3><center>$a_k U_{k-1} + b_k U_{k} + c_k U_{k+1}= f_k$ $(2)$</center></h3>

где   −  $a_k, b_k, c_k$ - некоторые коэффициенты.  Уравнения $(1)$ и $(2)$ будем называть соответственно линейными разностными уравнениями первого и второго порядка (при условии, что $a_k, b_k \ne 0$ в $(1)$ и$a_k, c_k \ne 0$ в $(2)$). Заметим, что при переходе от дифференциального уравнения к разностному уравнению, порядок уравнения не обязательно сохраняется.
Рассмотрим задачу вида:

<h3><center>$\begin{cases} a_k U_{k-1} + b_k U_k + c_k U_{k+1} = f_k, k = \overline{1, n-1}\\ U_0 = \phi , U_k = \psi \end{cases}$</center></h3>

называемую разностной краевой задачей (РКЗ). Данную задачу можно коротко переписать в виде:

<h3><center>$\begin{cases} LU_k = f_k, k = \overline{1, n-1}\\ U_0 = \phi , U_k = \psi \end{cases}$</center></h3>

где линейный разностный оператор $LU_k$ имеет вид:

<h3><center>$LU_k = a_k U_{k-1} + b_k U_k + c_k U_{k+1}$</center></h3>

Задачу РКЗ будем называть хорошо обусловленной, если она имеет решение, причем единственное, при любых значениях , и, кроме того, это решение ${Uk}$ удовлетворяет соотношению:

<h3><center>$ |U_k | \leq max(|\phi|, |\psi|, \underset{i}{max} |f_i|)$</center></h3>

где $M = const$ ,   не зависящая от  n.  


#### Теорема 1 (достаточный признак хорошей обусловленности). 
Пусть выполнимо следующее условие:

<h3><center>$|b_k| \geq |a_k| + |c_k| + \delta , \delta > 0$</center></h3>

Тогда задача РКЗ хорошо обусловлена, а условие хорошей обусловленности имеет следующий вид:

<h3><center>$ |U_k | \leq max(|\phi|, |\psi|, \frac{1}{\delta} \underset{i}{max} |f_i|)$</center></h3>

#### Теорема 2 (критерий хорошей обусловленности). 

Пусть $a+k = a, b_k = b, c_k = c$. Тогда задача является хорошо обусловленной тогда и только тогда, когда корни $q_1$, $q_2$ характеристического уравнения удовлетворяют условиям $|q_1|<1, |q_2|>1$ .


Разностная схема:

<h3>$L_h u^{(h)} = f^{(h)}$</h3>

называется устойчивой, если существует число $h_0 > 0$ такое, что при любом $h < h_0$ разностная задача имеет единственное решение при любой правой части f^{(h)}, и это решение удовлетворяет соотношению:

где $С$ – независимая от шага $h$ константа.

<h2><center>Метод прогонки.</center></h2>

Так как РКЗ представляет собой линейную трехдиагональную систему из $(n+1)$ уравнений с $(n+1)$ неизвестными, то для ее решения можно применить метод прогонки. Он обладает рядом преимуществ.

Достоинства:

+ число арифметических операций при использовании данного метода составляет $O(n)$, т. е. не превосходит $K_n$  , где  $K = const$.
+ при выполнении условия преобладания диагональных элементов $|b_k| \leq |c_k| + |a_k| + \delta$ метод прогонки оказывается слабо чувствительным к ошибкам вычислений: вычислительная погрешность не накапливается с ростом числа узлов $n$.

```python
import numpy as np
import sympy as sp
from sympy import cos, sin, exp
import math
from matplotlib import pyplot as plt
```

```python
def solve_diag(A, B, C, F):
    a = [0]
    b = [0]
    n = len(A)

    for i in range(n):
        a.append(-C[i] / (A[i] * a[i] + B[i]))
        b.append((F[i] - A[i] * b[i]) / (A[i] * a[i] + B[i]))

    y = np.zeros(n)
    y[n - 1] = (F[n - 1] - A[n - 1] * b[n - 1]) / (C[n - 1] + A[n - 1] * a[n - 1])
    for i in range(n - 2, -1, -1):
        y[i] = a[i + 2] * y[i + 1] + b[i + 2]

    return y
```

### Задание 1

Составить разностную схему и получить численное решение краевой задачи

$ay'' + (1 + bx^2)y = -1, -1<x<1$

с точностью $10^{-3}$. 
Исходные данные: $a=sin(k), b=cos(k)$, где $k$ номер варианта $(k=9)$.
Граничные условия выбрать однородными: $\begin{cases} y(-1) = 0 \\ y(1) = 0 \end{cases}$


```python
# Builds coefficients for diff equation in format y'' - p(x)y = f(x)
def build_coeffs_tridiag(x_start, x_end, h, p, g):
    x = sp.symbols('x')

    A, B, C, F = [], [], [], []
    for xi in np.arange(x_start, x_end + h, h):
        A.append(1)
        C.append(1)
        B.append(-h ** 2 * p.subs({x: xi}) - 2)
        F.append(h ** 2 * g(xi))

    return A, B, C, F
```

```python
def solve_task1(e=0.001):
    var = int(input("Enter variant: "))
    a = math.sin(var)
    b = math.cos(var)
    p, x = sp.symbols("p x")
    p = -(1 + b * x ** 2) / a

    x_start = -1
    x_end = 1

    h = 0.25
    cur_precision = 1

    print('Solving task1: ')
    print("y'' + (" + str(p) + ') * y = ' + str(-1 / a))
    print('q = ' + str(p))
    print('p = 0')
    print('x0 = ' + str(x_start))
    print('xn = ' + str(x_end))
    print('Start step = ' + str(h))

    A, B, C, F = build_coeffs_tridiag(x_start, x_end, h * 2, p, f(a))
    y1 = solve_diag(A, B, C, F)
    y = []

    while cur_precision > e:
        y = np.copy(y1)
        h /= 2
        A, B, C, F = build_coeffs_tridiag(x_start, x_end, h, p, f(a))
        y1 = solve_diag(A, B, C, F)
        cur_precision = np.max(np.abs([y1[2 * i] - y[i] for i in range(len(y))]))
        print('Precision: {}, Current step: {}'.format(cur_precision, h))

    x1 = np.arange(x_start, x_end + h, h)
    plt.plot(x1, y1)
    plt.show()
```

#### Решение
Составим разностную схему:

Заменив в исходном уравнении производные на разностные отношения

<h3>$y'(x_k) = \frac{y_{k+1} - y_k}{2h}, y'' = \frac{y_{k+1} - 2y_k + y_{k-1}}{h^2}, k = 1..n-1$</h3>

получим:

<h3>$y_{k-1} - (h^2 \frac{1+bx_k}{a} +2)y_k + y_{k+1} = \frac{-h^2}{a}, k =1..n, y_0=0, y_n=0$</h3>

Решить полученную систему уравнений с трехдиагональной матрицей можно с помощью метода прогонки.

Найдем с помощью метода прогонки решение данной системы и построим график полученной функции при необходимой точности

```python
solve_task1()
```

$7.62939453125e-06$

Как видим, при начальном шаге в $h_0=0.25$ понадобилось $15$ итераций, чтобы добиться необходимой точности.

### Задание 2

Методом конечных разностей найти приближенное решение краевой задачи:

<h3>$y'' - 4xy' + 2x^2y = 2x, 2\leq x \leq4$</h3>

c точностью $ε=0.07$ и построить его график. Решение системы разностных уравнений найти, используя метод прогонки.
Граничные условия: <h3>$\begin{cases} y(2) = 0 \\ y(4) -3y'(4)= 0 \end{cases}$</h3>

```python
def build_coeffs1(x_start, x_end, h, p, q, g):
    x = sp.symbols('x')
    n = int((x_end - x_start) / h) + 1

    a = np.zeros((n, n))
    b = np.zeros(n)
    for i, xi in enumerate(np.arange(x_start + h, x_end, h)):
        j = i + 1
        a[j][j - 1] = 2
        a[j][j] = 2 * h ** 2 * q.subs({x: xi}) - h * p.subs({x: xi}) - 4
        a[j][j + 1] = 2 + h * p.subs({x: xi})
        b[j] = 2 * h ** 2 * g.subs({x: xi})

    a[0][0] = -3
    a[0][1] = 4
    a[0][2] = -1
    b[0] = 0

    a[n - 1][n - 1] = 2 * h - 3
    a[n - 1][n - 2] = 12
    a[n - 1][n - 3] = -3
    b[n - 1] = 4 * h

    return np.linalg.solve(a, b)
```

```python
def solve_task2(e=0.07):
    p, q, g, x = sp.symbols("p q g x")
    p = -4 * x
    q = 5 + 0 * x
    g = 2 * x

    x_start = 2
    x_end = 4
    h = 0.1

    print('Solving task 2: ')
    print("y'' + (" + str(p) + ") * y' + (" + str(q) + ') * y = ' + str(g))
    print('q = ' + str(p))
    print('p = ' + str(q))
    print('x0 = ' + str(x_start))
    print('xn = ' + str(x_end))
    print('Start step = ' + str(h))

    cur_precision = 1
    y1 = build_coeffs1(x_start, x_end, h, p, q, g)
    y = []

    while cur_precision > e:
        y = np.copy(y1)
        h /= 2
        y1 = build_coeffs1(x_start, x_end, h, p, q, g)
        cur_precision = np.max(np.abs([y1[2 * i] - y[i] for i in range(len(y))]))
        print('Precision: {}, Current step: {}'.format(cur_precision, h))

    plt.plot(np.arange(x_start, x_end + h, h), y1)
    plt.show()
```

#### Решение



Заменив в исходном уравнении производные на разностные отношения

<h3>$y'(x_k) = \frac{y_{k+1} - y_k}{2h}, y'' = \frac{y_{k+1} - 2y_k + y_{k-1}}{h^2}, k = 1..n-1$</h3>

получим:

<h3>$\begin{cases} y_0' = \frac{-3y_0 + 4y_1 - y_2}{2h} \\
y_n' = \frac{3y_n - 4y_{n-1} + y_{n-2}}{2h}\end{cases}$</h3>

Решить полученную систему уравнений с трехдиагональной матрицей можно с помощью метода прогонки.

Найдем с помощью метода прогонки решение данной системы и построим график полученной функции при необходимой точности

### Задание 3

Методом конечных разностей найти приближенное решение краевой задачи с тремя верными значащими цифрами. Решение системы разностных уравнений:

<h3><center>$\begin{cases} -(k(x)u')' + q(x)u = f(x), x \in (a,b) \\
-k(a)u'(a) + 0.5u(a) = 0, \\
k(b)u'(b) + 0.5u(b) = 0\end{cases}$</center></h3>

найти, используя метод прогонки. 

<h3>$a=0, b=1.7, c=0.925$</h3>
<h3>$k(x)=1.8a<x<c$;</h3>
<h3>$k(x)=0.4c<x<b$;</h3>
<h3>$q(x)=7.0a<x<c$;</h3>
<h3>$q(x)=12.8c<x<b$;</h3>
<h3>$f(x)=\frac{8x}{2 + x^3}$</h3>


```python
def build_coeffs_tridiag_task4(x_start, x_end, h, p, g):
    x = sp.symbols('x')

    A, B, C, F = [], [], [], []
    for xi in np.arange(x_start, x_end + h, h):
        A.append(1)
        C.append(1)
        B.append(-h ** 2 * p(xi) - 2)
        F.append(h ** 2 * g(xi))

    A[0] = 3.6 + h
    B[0] = -4.8
    C[0] = 1.2
    F[0] = F[-1] = 0
    C[-1] = h + 1.2
    B[-1] = -1.5
    A[-1] = 0.4

    return A, B, C, F
```

```python
def solve_task3():
    p, x = sp.symbols("p x")

    k1 = 1.8
    k2 = 0.4
    q1 = 7.0
    q2 = 12.8
    g = 8 * x / (2 +  x ** 3)
    c = 0.925
    k = lambda xi: k1 if xi < c else k2
    q = lambda xi: q1 if xi < c else q2

    x_start = 0
    x_end = 1.7

    h = 0.25
    cur_precision = 1
    e = 0.05

    print('Solving task 3: ')
    print("y'' + q(x) / k(x) * y = " + str(g) + "/ k(x)")
    print('q = q(x) / k(x)')
    print('p = 0')
    print('x0 = ' + str(x_start))
    print('xn = ' + str(x_end))
    print('Start step = ' + str(h))

    A, B, C, F = build_coeffs_tridiag_task4(x_start,
                                            x_end,
                                            h * 2,
                                            lambda xi: -q(xi) / k(xi),
                                            lambda xi: -g.subs({x: xi}) / k(xi))

    y1 = solve_diag(A, B, C, F)
    y = []

    while cur_precision > e:
        y = np.copy(y1)
        h /= 2
        A, B, C, F = build_coeffs_tridiag_task4(x_start,
                                                x_end,
                                                h,
                                                lambda xi: -q(xi / k(xi)),
                                                lambda xi: -g.subs({x: xi}) / k(xi))
        y1 = solve_diag(A, B, C, F)
        cur_precision = np.max(np.abs([y1[2 * i] - y[i] for i in range(len(y) - 1)]))
        print('Precision: {}, Current step: {}'.format(cur_precision, h))

    x1 = np.arange(x_start, x_end + h, h)
    plt.plot(x1, y1)
    plt.show()
```
